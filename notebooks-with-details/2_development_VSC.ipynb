## Containerise the web application in Visual Studio Code

In a new folder in Visual Studio Code, I create a new ASP.NET WebAPI project by running two dotnet commands:

````python
dotnet new webapi --no-https # Configures a new ASP.NET Core Web API project without enabling HTTPS
dotnet build # Compiles the source code and generates the necessary executable or library files
````

Source code builds in less than 5 seconds! 



I then use Command Palette to run the following command: `.NET: Generate Assets for Build and Debug`

Add the following files into .gitignore file: 
[Bb]in/
[Oo]bj/

Finally, I can save all files in the `File` menu and then in `Source Control`, choose `Publish to Github` public repository. 




## Create Docker image and push to Azure Container Registry

I now create a Docker image: 

In the Command Palette, I run `Add Docker Files to Workspace` and chose the following options:
`.NET ASP.NET Core.` for Application Platform
`Linux` for Operating System
`5000` for Ports 
`No` for Docker Compose files

I then run the command in the terminal to build a container image hosted by Docker: 

```bash
docker build --tag aspnetcorecontainer:latest .

``` 
Then, I go back to the Command Palette and run `Docker Images: Push` selecting these options:
`aspnetcorecontainer` for Docker image
`latest` for image tag

I then connect to Azure Container Registry and press `Enter` to push the image into the chosen registry.

I then commit and sync changes in `Source Control` view.
