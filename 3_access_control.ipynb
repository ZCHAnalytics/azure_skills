## Configure Azure Container Registry for a secure connection with Azure Container Apps

Now in our Azure resource group I have a container registry instance with one image, a Virtual Network with subnets and a Service Bus Namespace. I need to configure a secure connection for the registry and container apps.

First, I create a User Assigned Managed Identity named `uai-apl2003`:

Then, in `Access Control (IAM)` in Container Registry, I add a role assignment `AcrPull`' and choose `User-assigned managed identity` for `Managed identity`.

In the Container Registry, I open `Networking`, choose `Private access` and add a new private endpoint connection named `pe-acr-apl2003`
