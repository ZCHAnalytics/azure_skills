## 1. Set revision management to multiple

IN  Application's Revisions and replicas, swtich from single to multi-revision mode.
The Revision Mode will be set to Multiple after the update.

## 2. Create a new revision with a v2 suffix

Create new revision, add version number and choose a container image.


## 3. Configure labels on the revisions

Enbale Ingress under Settings and specify the following information:

- Ingress traffic: Accepting traffic from anywhere
- Ingress type: HTTP
- Client certificate mode: Ignore
- Transport: Auto
- Insecure connections: Allowed NOT checked
- Target port: 5000
- IP Security Restrictions Mode: Allow all traffic

Under Revisions, Revisions and replicas, update labels for:
- the v2 revision as `updated`
- other revision as `current`

## 4. Configure a traffic percentage on the revisions

In Traffic under Revisions and replicas add: 
- 25 as the percentage for the v2 revision
- 75 as the percentage for the other revision 


## 5. Verify the configuration
