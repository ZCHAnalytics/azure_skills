## 1. Configure Pipeline1 to use the self-hosted agent pool

Edit project pipeline and commit and save

## 2. Configure Pipeline1 with an `Azure Container Apps Deploy` task

From `Tasks` select `Azure Container Apps Deploy`

- Azure Resource Manager > Authorize the subscription use
- Docker Image to Deploy: `Registry`/`Image`:`Image tag`
- Azure Container App name: `Name`
- Azure Resource group name: `RG1` 
- Save and commit
