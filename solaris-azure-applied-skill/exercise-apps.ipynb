# Create and configure a container app in Azure Container Apps

## Tasks:
## 1. Create a container app that uses an ACR image
## 2. Configure the container app to authenticate using the user assigned identity
## 3. Configure a connection between the container app and Service Bus
## 4. Configure HTTP scale rules
## 5. Verify the configuration

## 1. Create a container app that uses an ACR image

- Create Container App named aca-apl2003
- Create a Container Apps Environment
- In Environment Networking tab, select your own virtual network, VNET1, ACASubnet (maybe need to adjust the address range)

In Container tab, uncheck `Use quickstart image`
- Ensure that Azure Container Registry is selected as image source.
- Select your container registry acrapl2003cah.azurecr.io
- Select image aspnetcorecontainer
- select Image tag: latest

## 2. Configure the container app to authenticate using the user assigned identity

Add >- Container App > Identity > user assigned identity 


## 3. Configure a connection between the container app and Service Bus

- In Settings' Service Connector (Preview), Create connection to Service Bus with Client type `.NET` with authentication `User assigned managed identity`

## 4. Configure HTTP scale rules

- In Application's Revisions and replicas > Scale > Edit and deploy > Scale
- Min replicas: 0; and Max replicas: 2
- Add Scale rule named scalerule-http, type: HTTP scaling and concurrent requests: 10,000.

## 5. Verify the configuration

az containerapp connection show --connection <service bus name> --name aca-apl2003 --resource-group solaris-rg1
